```python
import pandas as pd
```

```python
iris_csv=pd.read_csv('/content/iris.csv')
iris_excel=pd.read_excel('/content/iris.xlsx',sheet_name='iris')
def explore_dataset(dataset):
  print('First five rows: ')
  print(dataset.head(),"\n\n")
  print('Last five rows: ')
  print(dataset.tail(),"\n\n")
  print('Stats: ')
  print(dataset.describe(),"\n\n")
  print('Information: ')
  print(dataset.info(),"\n\n")
print('FOR CSV FILE: \n\n')
explore_dataset(iris_csv)
print('FOR EXCEL FILE: \n\n')
explore_dataset(iris_excel)
```
